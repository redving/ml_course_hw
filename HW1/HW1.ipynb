```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
#загружаю датасет
data = pd.read_csv('adult.csv')
```

```python
#смотрю данные
data.head()
```

```python
#проверяем наличие Null
data.info()
```

```python
#перевожу целевую переменную
le = LabelEncoder()
le.fit(data['income'])
y = pd.Series(le.transform(data['income']))
```

```python
#делаем onehotencoding
selected_columns = ['workclass','marital-status','occupation','relationship','race','gender','native-country']
X = pd.get_dummies(data, columns = selected_columns, drop_first = True) 
```

```python
#удаляем дубль
del X['education']
#удаляем целевую переменную
del X['income']

del X['fnlwgt']
```

```python
X.head()
```

```python
#делаем стандартизацию
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 999)
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_prediction = lr_model.predict(X_test)
```

```python
lr_model.score(X_test, y_test)
```

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_prediction = svm_model.predict(X_test)
```

```python
svm_model.score(X_test, y_test)
```
